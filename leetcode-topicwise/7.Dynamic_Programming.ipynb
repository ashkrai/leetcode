```python
# Fibonacci Number
# Problem: https://leetcode.com/problems/fibonacci-number/description/

def fib(n):
    """
    Compute the nth Fibonacci number using dynamic programming.
    """
    if n <= 1:
        return n
    dp = [0] * (n + 1)
    dp[1] = 1
    for i in range(2, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]
    return dp[n]

# Sample Input
n = 10

# Compute the nth Fibonacci number
print(fib(n))  # Expected Output: 55

# Time Complexity: O(n)
# Space Complexity: O(n)
# The dynamic programming approach computes each Fibonacci number once, resulting in linear time complexity and space complexity.
```

```python
# Climbing Stairs
# Problem: https://leetcode.com/problems/climbing-stairs/description/

def climb_stairs(n):
    """
    Calculate the number of ways to reach the nth step.
    """
    if n <= 2:
        return n
    dp = [0] * (n + 1)
    dp[1], dp[2] = 1, 2
    for i in range(3, n + 1):
        dp[i] = dp[i - 1] + dp[i - 2]
    return dp[n]

# Sample Input
n = 5

# Calculate the number of ways to reach the nth step
print(climb_stairs(n))  # Expected Output: 8

# Time Complexity: O(n)
# Space Complexity: O(n)
# The dynamic programming approach calculates the number of ways to reach each step, resulting in linear time and space complexity.
```

```python
# Longest Increasing Subsequence
# Problem: https://leetcode.com/problems/longest-increasing-subsequence/description/

def length_of_lis(nums):
    """
    Find the length of the longest increasing subsequence.
    """
    if not nums:
        return 0
    dp = [1] * len(nums)
    for i in range(1, len(nums)):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)

# Sample Input
nums = [10, 9, 2, 5, 3, 7, 101, 18]

# Find the length of the longest increasing subsequence
print(length_of_lis(nums))  # Expected Output: 4

# Time Complexity: O(nÂ²)
# Space Complexity: O(n)
# The dynamic programming approach compares each pair of elements, resulting in quadratic time complexity and linear space complexity.
```

```python
# Maximum Subarray Sum
# Problem: https://leetcode.com/problems/maximum-subarray/description/

def max_sub_array(nums):
    """
    Find the contiguous subarray with the largest sum.
    """
    max_sum = current_sum = nums[0]
    for num in nums[1:]:
        current_sum = max(num, current_sum + num)
        max_sum = max(max_sum, current_sum)
    return max_sum

# Sample Input
nums = [-2, 1, -3, 4, -1, 2, 1, -5, 4]

# Find the maximum subarray sum
print(max_sub_array(nums))  # Expected Output: 6

# Time Complexity: O(n)
# Space Complexity: O(1)
# The dynamic programming approach iterates through the array once, maintaining the current sum and maximum sum,
#    resulting in linear time complexity and constant space complexity.
```

```python
# Coin Change
# Problem: https://leetcode.com/problems/coin-change/description/

def coin_change(coins, amount):
    """
    Find the fewest number of coins needed to make up the given amount.
    """
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1

# Sample Input
coins = [1, 2, 5]
amount = 11

# Find the fewest number of coins needed
print(coin_change(coins, amount))  # Expected Output: 3

# Time Complexity: O(n * m)
# Space Complexity: O(n)
# The dynamic programming approach iterates through each coin and each amount, resulting in time complexity proportional
#    to the product of the number of coins and the amount.
```
